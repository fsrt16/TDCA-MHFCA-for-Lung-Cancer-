```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
models = [
    "VGG16", "VGG19", "ResNet50", "DenseNet121", "EfficientNetV1-B0", "InceptionV3", "InceptionV4",
    "MobileNetV3", "MobileViT-V2", "Xception", "Inception-Next", "MaxViT", "Swin Transformer", "DeiT3",
    "ResMLP", "GhostNet", "FastViT", "CaFormer", "RepGhostNet", "Exception (Base)", "Proposed - TDCA-MHFCA"
]

accuracy = [
    0.887, 0.892, 0.910, 0.915, 0.921, 0.902, 0.912,
    0.891, 0.912, 0.893, 0.925, 0.926, 0.935, 0.918,
    0.899, 0.890, 0.929, 0.931, 0.892, 0.942, 0.9818
]

f1_score = [
    0.870, 0.876, 0.888, 0.893, 0.899, 0.882, 0.890,
    0.873, 0.887, 0.879, 0.902, 0.907, 0.918, 0.895,
    0.880, 0.872, 0.910, 0.912, 0.875, 0.944, 0.9804
]

weighted_precision = [
    0.875, 0.881, 0.892, 0.898, 0.903, 0.887, 0.894,
    0.878, 0.891, 0.884, 0.908, 0.911, 0.923, 0.899,
    0.885, 0.877, 0.915, 0.917, 0.880, 0.948, 0.9825
]

weighted_recall = [
    0.867, 0.874, 0.885, 0.890, 0.896, 0.879, 0.888,
    0.870, 0.885, 0.876, 0.899, 0.904, 0.916, 0.892,
    0.877, 0.869, 0.908, 0.911, 0.872, 0.943, 0.9818
]


# Plotting function with shared legend
def plot_subplots(models, accuracy, f1_score):
    sns.set_style("whitegrid")
    markers = ['o', 's', 'D', '^', 'v', '<', '>', 'p', '*', 'h', 'H', '+', 'x']
    colors = sns.color_palette("Set2", n_colors=len(models))

    fig, axes = plt.subplots(1, 2, figsize=(20, 8), sharey=True)

    metrics = [accuracy, f1_score]
    titles = ["Accuracy", "F1-Score"]

    for ax, metric, title in zip(axes, metrics, titles):
        ax.plot(models, metric, linestyle='dashed', color='gray', alpha=0.6, marker='', markersize=8)
        for i, (model, value) in enumerate(zip(models, metric)):
            ax.scatter(
                model, value,
                marker=markers[i % len(markers)],
                s=120, color=colors[i],
                edgecolor='black', label=model
            )

        ax.set_title(f"{title} Comparison", fontsize=16)
        ax.set_ylabel(f"{title} (%)", fontsize=14)
        ax.tick_params(axis='x', rotation=90, labelsize=10)
        ax.tick_params(axis='y', labelsize=12)

    # Create a single legend
    handles, labels = axes[0].get_legend_handles_labels()
    fig.legend(
        handles, labels, loc='lower center', bbox_to_anchor=(0.15, -0.05), 
        ncol=6, fontsize=10, title="State of the Art - Comparitive Experimental Results",
        title_fontsize='13', frameon=True, shadow=True, fancybox=True
    )

    plt.tight_layout(rect=[0, 0.05, 1, 1])  # Leave space for legend
    plt.show()

# Call the function
plot_subplots(models, accuracy, f1_score)
```
